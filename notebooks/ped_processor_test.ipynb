```python
from src.processing.PreprocessingEducationalData import PreprocessingEducationalData
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename_processed = '../data/processed/StudentPerformanceFactors.csv'
df = pd.read_csv(filename_processed)

df.head()
```

```python
preprocessor = PreprocessingEducationalData()
```

```python
"""
Inspecciona las transformaciones aplicadas a los datos
"""
processed_df = preprocessor.fit_transform(df)

print("\nResumen de características:")
print(f"Número total de características: {len(processed_df.columns)}")

print("\nVariables numéricas (estandarizadas):")
for feature in preprocessor.numeric_features:
    if feature in processed_df.columns:
        print(f"- {feature}: Media = {processed_df[feature].mean():.3f}, "
                f"Std = {processed_df[feature].std():.3f}")

print("\nVariables nominales (one-hot):")
nominal_cols = [col for col in processed_df.columns 
                if any(feat in col for feat in preprocessor.nominal_features)]
for col in nominal_cols:
    print(f"- {col}: Valores únicos = {processed_df[col].unique()}")

print("\nVariables ordinales (label encoded):")
for feature in preprocessor.ordinal_mappings.keys():
    if feature in processed_df.columns:
        print(f"- {feature}: Valores únicos = {sorted(processed_df[feature].unique())}")
        if feature == 'Peer_Influence':
            print(f"  Codificación: {preprocessor.ordinal_mappings[feature]}")
```
